```python
# Negative transformation of the image

import cv2
import matplotlib.pyplot as plt

# Read Image
img_bgr = cv2.imread('mustang-gt.jpg', 1)

# Display Image
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.title("Original Image")
plt.axis('off')
plt.show()

# Print pixel grid values (top-left 5x5)
print("Top-left 5x5 pixel grid (BGR values):\n")

rows, cols = 5, 5
for i in range(rows):
    for j in range(cols):
        print(f"Pixel[{i},{j}] = {img_bgr[i, j]}")
    print()

# Negative image transformation
img_negative = 255 - img_bgr

# Display negative image
img_negative_rgb = cv2.cvtColor(img_negative, cv2.COLOR_BGR2RGB)
plt.imshow(img_negative_rgb)
plt.title("Negative Image")
plt.axis('off')
plt.show()

```

```python
import cv2
import matplotlib.pyplot as plt

# Contrast Streching
stretched_image = cv2.normalize(img_bgr, None, 0, 255, cv2.NORM_MINMAX)

# Display the results
stretched_image_rgb = cv2.cvtColor(stretched_image, cv2.COLOR_BGR2RGB)
plt.imshow(stretched_image_rgb)
plt.title("Contrast Stretched Image")
plt.axis('off')
plt.show()

print("Top-left 5x5 pixel grid (BGR values) for Contrast Stretched Image:\n")
rows, cols = 5, 5
for i in range(rows):
    for j in range(cols):
        print(f"Pixel[{i},{j}] = {stretched_image[i, j]}")
    print()
```

```python
 # Convert image to grayscale for simplicity
img_gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)

# Find min and max pixel values
I_min = img_gray.min()
I_max = img_gray.max()

print(f"Minimum pixel value: {I_min}")
print(f"Maximum pixel value: {I_max}")

# Apply contrast stretching
img_contrast = ((img_gray - I_min) / (I_max - I_min) * 255).astype('uint8')

# Display contrast-stretched image
plt.imshow(img_contrast, cmap='gray')
plt.title("Contrast Stretched Image")
plt.axis('off')
plt.show()

# Print pixel grid values (top-left 5x5) AFTER contrast stretching
print("Top-left 5x5 pixel grid (after contrast stretching):\n")

rows, cols = 5, 5
for i in range(rows):
    for j in range(cols):
        print(f"Pixel[{i},{j}] = {img_contrast[i, j]}")
    print()
```

```python
# Convert image to grayscale
imgGT = cv2.imread('Rohit.jpg', 1)
img_gray2 = cv2.cvtColor(imgGT, cv2.COLOR_BGR2GRAY)

# Display original grayscale image
plt.imshow(img_gray2, cmap='gray')
plt.title("Grayscale Image")
plt.axis('off')
plt.show()

# Extract 8 bit planes (0 to 7)
bit_planes = []

for k in range(8):
    bit_plane = (img_gray2 >> k) & 1
    bit_planes.append(bit_plane)

    # Display each bit plane
    plt.figure(figsize=(3, 3))
    plt.imshow(bit_plane * 255, cmap='gray')
    plt.title(f"Bit Plane {k}")
    plt.axis('off')
    plt.show()


print("Top-left 5x5 pixel grid (Bit Plane 7 - MSB):\n")

rows, cols = 5, 5
for i in range(rows):
    for j in range(cols):
        print(f"Pixel[{i},{j}] = {bit_planes[7][i, j]}")
    print()
```
