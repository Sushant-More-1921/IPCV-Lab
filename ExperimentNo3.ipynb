```python
# Enhancement-of-Image-using-Histogram
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Read Image
img_rgb = cv2.imread('/content/img2.webp', cv2.IMREAD_GRAYSCALE)
img_rgb1 = cv2.imread('/content/img2.webp', cv2.IMREAD_GRAYSCALE)
plt.imshow(img_rgb,cmap='gray')
plt.title("Original Image")
plt.axis('off')
plt.show()
```

```python
# initialize an array hist with zeros of size 256
hist = np.zeros(256)
```

```python
#for each pixel value p in img increament hist[p] by 1
for p in img_rgb.flatten():
    hist[p] += 1
```

```python
#Normallize hist by dividing each value by total number of pixels
hist = hist / img_rgb.size

hist
```

```python
# compute cummulative sum of hist to obtain cdf
cdf = np.cumsum(hist)

cdf
```

```python
#for each gray level i from 0 to 255 New_value[i] = round(255 * cdf[i])
for i in range(256):
    cdf[i] = np.uint8(255 * cdf[i])
```

```python
#for each pixel p in img : replace p with new_value[p]
for i in range(img_rgb.shape[0]):
    for j in range(img_rgb.shape[1]):
      img_rgb[i][j] = cdf[img_rgb[i][j]]
```

```python
#Display the histogram equalized image

#original image
plt.imshow(img_rgb1,cmap='gray')
plt.title("Original Image")
plt.axis('off')
plt.show()

plt.imshow(img_rgb,cmap='gray')
plt.title("Histogram Equalized Image")
plt.axis('off')
plt.show()
```
